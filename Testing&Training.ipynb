```python
def return_clean_text(comments):
    import string
    clean_text=[]
    comments=[comments]
    for i in comments:
        texts=i.split(' ')
        arr=[]
        for j in texts:
            j=j.translate(str.maketrans('', '', string.punctuation))
            j=''.join([i for i in j if not i.isdigit()])
            j=j.lower()
            if "\n" in j:
                brr=j.split('\n')
                arr+=brr
            else:
                arr.append(j)
        clean_text.append(arr)

    return clean_text[0]+[-1]*(200-len(clean_text[0])) if len(clean_text[0])<=200 else clean_text[:200]
```

```python
u=hashed["hash_list_reverse"]
v=hashed["hash_list"]

def tokenizeComment(comments):
    clean_text=return_clean_text(comments)
    solution=[]
    for i in clean_text:
        if i in v:
            solution.append(v[i])
        else:
            solution.append(-1)
    return solution

def predict_tokenize(comments):
    tokens=tokenizeComment(comments)
    token_pred=clf.predict([tokens])
    return token_pred
```
